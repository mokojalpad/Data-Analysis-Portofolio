### Cleaning Data

#### Drop Unnecessary Column

```python
# drop columns
cols_drop=["n_achievements", "additional_content", "required_age", "review_score_desc"]
df.drop(columns=[col for col in cols_drop if col in df.columns ], inplace=True)
```

#### Clean Date Column

```python
pd.options.display.max_rows=None
display(df["release_date"].value_counts())
```

```python
# Replace value "Not Released" to NaT (Not a Time)
df["release_date"] = df["release_date"].replace("Not Released", pd.NaT)
```

```python
# checking it
df["release_date"].unique()
```

```python
# remove time in date value
df["release_date"] = df["release_date"].str.split().str[0]
```

```python
# checking it
display(df["release_date"].unique())
```

```python
# convert "release_date" to datetime
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")
```

```python
# checking it
df["release_date"].unique()
```

#### Clean Categories Column

```python
# check unique values
df["categories"].unique()
```

```python
# check type of 1st row value
type(df["categories"][0])
```

```python
# Convert string to list for next step 
df["categories"] = df["categories"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function for set values based on "categories"
# Single-player = 2
# Multi-player = 3
# Single-player & Multi-player = 4
# Else = 1
def assign_mode_value(categories):
    categories_set = set(categories)
    if "Single-player" in categories_set and "Multi-player" in categories_set:
        return "Single-Multi-Player"
    elif "Single-player" in categories_set:
        return "Single-player"
    elif "Multi-player" in categories_set:
        return "Multi-player"
    else:
        return "Unknown"

df["game_mode_value"] = df["categories"].apply(assign_mode_value)
```

```python
df = df.drop(columns=["categories"])
```

```python
# checking it
df["game_mode_value"].value_counts()
```

#### Clean Genres Column

```python
df["genres"].unique()
```

```python
# Convert str to list
df["genres"] = df["genres"].apply(ast.literal_eval)

# Expand "genres" to different rows
df = df.explode("genres")

# reset index for no duplicate index
df.reset_index(drop=True, inplace=True)

df.head()
```

```python
df.head()
```

```python
# fill no value with No Genres
df["genres"] = df["genres"].fillna("No Genres")
df["genres"].isna().sum()
```

#### Clean Is_released Column

```python
df["is_released"] = df["is_released"].map({True: "Released", False: "Not Released"})

df.head()
```

#### Clean Total_positive & Total_negative

```python
# calculate ratio from "total_positive" to "total_negative"
df["review_ratio"] = df["total_positive"] / df["total_negative"].replace(0,1) # avoid 0 divided

# create rating system based ratio
def rating_system(ratio, total_reviews):
    if total_reviews == 0:
        return "Not Review Yet" 
    elif ratio > 2:
        return "Very Positive"
    elif 1 <= ratio <=2:
        return "Positive"
    elif 0.5 <= ratio <1:
        return "Neutral"
    else:
        return "Negative"

# apply rating_system to new column
df["rating_ratio"] = df.apply(lambda x: rating_system(x["review_ratio"], x["total_reviews"]), axis=1)

# drop total_positive", "total_negative", "positive_percentual" column
df = df.drop(columns=["total_positive", "total_negative", "positive_percentual"])

df.head(100)
```

```python
sorted(df["metacritic"].unique())
```

#### Clean Metacritic Column

```python
# create rating form metacritic
def class_metacritic(score):
    if score >= 0 and score <= 40:
        return "Negative"
    elif score >=41 and score <= 60:
        return "Neutral"
    elif score >= 61 and score <= 80:
        return "Positive"
    elif score >=81 and score <= 100:
        return "Very Positive"
    else:
        return "Invalid" # for give invalid value

# apply class_metacritic to new column 
df["metacritic_rating"] = df["metacritic"].apply(class_metacritic)

df.head(100)
```

#### Clean Is_free Column

```python
df["is_free"] = df["is_free"].map({True: "Free", False: "Paid"})

df.head()
```

#### Clean Price_initial (USD) Column

```python
sorted(df["price_initial (USD)"].unique())
```

```python
# create rating form metacritic
def class_price(price):
    if price == 0:
        return "Free"
    elif 0 < price <= 5:
        return "Cheap"
    elif 5 < price <= 20:
        return "Modarate"
    elif 20 < price <= 50:
        return "Expensive"
    else:
        return "Very Expensive" 

# apply class_metacritic to new column 
df["price_category"] = df["price_initial (USD)"].apply(class_price)

df.head(100)
```

#### Clean Name, Developers, Publishers & Platforms Columns

```python
df["name"] = df["name"].str.strip() # delete extra space
# df["name"] = df["name"].str.strip().replace({r"[^a-zA-Z0-9\s]": ""}, regex=True) # delete non alfanumeric char

# fill null value with No_Name_Game
df["name"] = df['name'].fillna("No_Name_Game")
```

```python
# convert list on str to list python with ast.literal_eval
def convert_to_list(value):
    try:
        # check values can change to list use ast.literal_eval
        if isinstance(value, str):
            # handle valid list, then join elemnts list to str
            parsed_value = ast.literal_eval(value)
            if isinstance(parsed_value, list):
                # change element on the list to str before join
                parsed_value = [str(item) for item in parsed_value]
                return ", ".join(parsed_value)
            else:
                return "Unknown" # if not list return Unknown
        else:
            return "Unknown" # for value not str, return Unknown
    except (ValueError, SyntaxError):
        return "Unknown" # if error return Unknown

# convert list to str with comma (,)
df["developers"] = df["developers"].apply(convert_to_list)
df["publishers"] = df["publishers"].apply(convert_to_list)

# delete extra space
df["developers"] = df["developers"].str.strip()
df["publishers"] = df["publishers"].str.strip()

# fill null value with
df["developers"] = df["developers"].fillna("No_Name_Developers")
df["publishers"] = df["publishers"].fillna("No_Name_Publishers")
```

```python
df["platforms"] = df["platforms"].apply(convert_to_list)
```
